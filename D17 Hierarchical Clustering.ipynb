```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn

# AgglomerativeClustering == bottom-up
```

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
# this is used into all the data points is b/w 0 and 1
# This input i is typically a column
# i.min(): This finds the minimum value in the input data i
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:])
df_norm
```

```python
# create dendrogram
# single ---> Nearest Neighbour
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
# create dendrogram
# Compete ---> Furether Neighbour
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create dendrogram

dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))


# Average , centroid are same --->
"""Average linkage calculates the distance between two clusters as the average distance between all pairs of data points,
where one data point is in one cluster and the other data point is in the other cluster.
The comment in the code suggests that the 'centroid' method is similar to the 'average' method,
but it is actually a distinct linkage method."""
```

```python
# create dendrogram
# Average , centroid are same --->
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='centroid'))
```

```python
# create clusters
# AgglomerativeClustering ---> bottom to top

hc = AgglomerativeClustering(n_clusters=4 , linkage = 'single')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])

# create a data frame y_hc
```
